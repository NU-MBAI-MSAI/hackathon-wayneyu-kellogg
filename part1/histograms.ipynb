```python
import matplotlib.pyplot as plt
import random
```

```python
def model_two(npts):
    numbers = []
    for _ in range(npts):
        numbers.append(random.random())
    plt.hist(numbers, 50)
    plt.show()
```

```python
 model_two(100)
```

```python
 model_two(1000)
```

```python
 model_two(10000)
```
